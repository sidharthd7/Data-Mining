Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Loading a CSV file

```python
df = pd.read_csv('winequality-white.csv', delimiter  = ';')  

df.head()
```

2. Identifying the type of each attribute

```python
print("\n Data Types of Each Column:")
print(df.dtypes)

print("\n Checking for Missing Values:")
print(df.isnull().sum())

print("\n Dataset Shape (Rows, Columns):", df.shape)
```

3. Describing various statistical parameters related to the attributes such as
mean, median, min value, max value, standard deviation etc.

```python
print("\n Summary Statistics:")
print(df.describe())

print("\n Unique Values Per Column:")
print(df.nunique())
```

4. Determine the number of samples belonging to each target class.

```python
# Binary Classification (1 = Good, 0 = Bad)
df['quality'] = np.where(df['quality'] >= 6, 1, 0)  

print("\n Class Distribution:")
print(df['quality'].value_counts())
```

5. Compute Covariance & Correlation between each pair of attributes.

```python
print("\n Covariance Matrix:")
print(df.cov())

print("\n Correlation Matrix:")
print(df.corr())
```

6. Perform Data Visualization

    a. Histogram

```python
df.hist(figsize=(12, 8), bins=20)
plt.suptitle("Histograms of Features", fontsize=16)
plt.show()
```

    b. Density Plots

```python
df.plot(kind='density', subplots=True, layout=(3, 4), figsize=(12, 8), sharex=False)
plt.suptitle("Density Plots of Features", fontsize=16)
plt.show()
```

    c. Correlation Matrix (Heatmap)

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap of Correlation Matrix", fontsize=16)
plt.show()
```

    d. Box Plot

```python
plt.figure(figsize=(12, 6))
df.boxplot()
plt.xticks(rotation=45)
plt.title("Box Plot of Features", fontsize=16)
plt.show()
```

    e. Scatter Plot 

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='alcohol', y='quality')
plt.title("Scatter Plot: Alcohol vs. Quality", fontsize=16)
plt.show()
```

    f. Line Plot

```python
plt.figure(figsize=(10, 5))
df['residual sugar'].plot(kind='line')
plt.title("Line Plot of Residual Sugar", fontsize=16)
plt.show()
```
